```python
from bs4 import BeautifulSoup
import requests

whatmobile = "https://www.whatmobile.com.pk/"
source = requests.get(whatmobile).text
soup = BeautifulSoup(source,"lxml")

# print(soup.prettify())

menu = soup.find('div',class_="verticalMenu")
section = menu.find_all('section')[1]

links = []
for li in section.ul.find_all('li'):
    links.append(li.a.get("href"))

total_mobiles = 0

import csv

with open("MobileDatabase.csv","w",newline = "") as f:
    handler = csv.writer(f,delimiter = ",")
    handler.writerow(['Model Name','RAM','ROM','CPU','OS','Price'])
    
    useless_companies = [10,17,21,22,23]
    for index,link in enumerate(links[:27]):
        if index in useless_companies:
            continue
        company = whatmobile + link
        print(company)
        source = requests.get(company).text
        soup = BeautifulSoup(source,"lxml")
        try:
            mobiles = soup.find('div',class_="mobiles")
        # print(mobiles.prettify())

            models = []
            for item in mobiles.find_all('div',class_="item"):
                link = item.div.a.get("href")
                models.append(link)

            for i in models[:5]:
                source = requests.get("https://www.whatmobile.com.pk" + i).text
                try:
                    soup = BeautifulSoup(source,'lxml')

                    specs = []
                    for td in soup.find_all('td',class_='fasla RowBG1 specs-value bottom-border'):
                        specs.append(td.text)

                    model = soup.find('h1',class_='hdng3').text
                    price = soup.find('td',class_='fasla RowBG1 bottom-border-section right-border specs-subHeading')
                    price = price.strong.text


                    os = specs[0].strip()
                    cpu = specs[2].strip()
                    mem = specs[4].split(",")
                    rom = mem[0].strip()
                    ram = mem[1].strip()
                    print("Model: ",model,"\nos: ",os,"\ncpu: ",cpu,"\nram: ",ram,"\nrom:",rom,"\nPrice: ",price)
                    handler.writerow([model,ram,rom,cpu,os,price])
#                     q = input("Do you want to Add Entry? y/n: ")
#                     if q == 'y':
#                         handler.writerow([model,ram,rom,cpu,os,price])
#                     else:
#                         continue
                except:
                    print("Format Not Supported")
                    
            
        except:
            models = []
            for td in soup.find_all('td',class_='BiggerText'):
                if td.a == None:
                    continue
                else:
                    link = td.a.get('href')
                    models.append(link)
            for i in models[:5]:
                source = requests.get("https://www.whatmobile.com.pk" + i).text
                try:
                    soup = BeautifulSoup(source,'lxml')

                    specs = []
                    for td in soup.find_all('td',class_='fasla RowBG1 specs-value bottom-border'):
                        specs.append(td.text)

                    model = soup.find('h1',class_='hdng3').text
                    price = soup.find('td',class_='fasla RowBG1 bottom-border-section right-border specs-subHeading')
                    price = price.strong.text


                    os = specs[0].strip()
                    cpu = specs[2].strip()
                    mem = specs[4].split(",")
                    rom = mem[0].strip()
                    ram = mem[1].strip()
                    print("Model: ",model,"\nos: ",os,"\ncpu: ",cpu,"\nram: ",ram,"\nrom:",rom,"\nPrice: ",price)
                    handler.writerow([model,ram,rom,cpu,os,price])
#                     q = input("Do you want to Add Entry? y/n: ")
#                     if q == 'y':
#                         handler.writerow([model,ram,rom,cpu,os,price])
#                     else:
#                         continue
                except:
                    print("Format Not Supported")   
print("The End")
```
